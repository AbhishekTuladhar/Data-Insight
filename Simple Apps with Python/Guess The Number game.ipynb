# 'Guess the number' game

```python
import random
num = random.randint(1, 100) #random numbers between 1 and 100

#inializing count to zero
count = 0
print("Guess the number. You will have 7 attempts to make the correct guess.")

#a loop to continue for 7 iteration
while count<7:
    guess = int(input("Guess a number: "))
    count += 1
    if guess == num:
        print("Correct guess. CONGRATULATIONS!!")
        print("The correct number was %d." %num)
        break    #break the loop after the correct guess
    elif guess < num:
        print("Too low.")
    else:
        print("Too high.")

#conditional to provide information after seven attempts are over
if guess != num:
    print("Sorry. The correct number was %d." %num)
    print("Better luck next time.")
```
